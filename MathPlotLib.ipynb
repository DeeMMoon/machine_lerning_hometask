```python
import matplotlib.pyplot as plt
import math
import numpy as np
import scipy.stats as sps
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=np.arange(-20,20,0.1)
plt.plot(x,x*x+2*x+6)
plt.grid(ls='--')
plt.show()
```

```python
t = np.linspace(0, 4 * np.pi, 100)
plt.figure()
plt.polar(t,np.sin(3 * t) + np.cos(5 * t) , '--r')
plt.grid(color='g',linestyle='--', linewidth=0.7)
plt.show()
```

```python
grid  = np.linspace(0,3,100)
plt.figure(figsize=(10,4))
plt.grid(ls='--')
sample = sps.expon.rvs(size=10)
plt.scatter(sample, np.zeros(10) - 0.02, alpha=0.3, label='sample',linewidths=3)


plt.plot(grid, sps.expon.pdf(grid), color='green', linewidth=4)

plt.grid(ls='--')

plt.show()
```

```python
plt.figure()
n, bins, patches = plt.hist(sample, range=(0,3), bins = 20, normed =True)
plt.plot(grid, sps.expon.pdf(grid), color='red', label='density')
plt.show()
```

```python
grid  = np.linspace(0,3,100)
plt.figure(figsize=(10,4))
plt.grid(ls='--')
sample = sps.expon.rvs(size=500)
plt.scatter(sample, np.zeros(500) - 0.02, alpha=0.3, label='sample',linewidths=3)


plt.plot(grid, sps.expon.pdf(grid), color='green', label='density', linewidth=4)

plt.grid(ls='--')

plt.show()
```

```python
n = 256
u = np.linspace(0, 1, n)

x, y = np.meshgrid(u, u)
z = np.zeros((n, n, 3))
z[:, :, 0] = x
z[:, :, 2] = np.exp(x)

plt.figure()
plt.imshow(z)
plt.show()
```

```python
n = 256
u = np.linspace(0, 1, n)

x, y = np.meshgrid(u, u);
z = np.zeros((n, n, 3))
z[:, :, 0] = x
z[:, :, 2] = np.sin(x)

plt.figure()
plt.imshow(z)
plt.show()
```

```python
t = np.linspace(0,2 * np.pi,50)
r=0.4
th, ph = np.meshgrid(t,t)
x, y, z = (1 + r * np.cos(ph)) * np.cos(th), (1 + r * np.cos(ph)) * np.sin(th), r * np.sin(ph)
fig=plt.figure()
ax=Axes3D(fig)
ax.elev = 70
ax.set_aspect(0.5)
ax.plot_surface(x,y,z, rstride=2, cstride=1, cmap='gnuplot')
plt.show();
```
