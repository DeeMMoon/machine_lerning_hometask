```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline

def f(x, y):
    '''Функция двух переменных'''
    return np.sin(x) * np.cos(y)

def draw_f(grid_x, grid_y):
    '''Функция отрисовки функции f(x, y)'''
   
    fig = plt.figure(figsize=(10, 8))
    ax = Axes3D(fig)
    ax.plot_surface(grid_x, grid_y, f(grid_x, grid_y), cmap='inferno')
    plt.show()

x=np.linspace(0,2 * np.pi,100)
y=np.linspace(0,2 * np.pi,100)
grid_x, grid_y = np.meshgrid(x,y)# Ваш код здесь

draw_f(grid_x, grid_y)
```

```python
def transformation(X,a=1):                      #Задача №3 решение c NumPy
    B=np.array(X.copy())
    
    for i in range(1,len(X)+1):
            if i%2==1:
              X[i-1]=a
            else:
                X[i-1]=X[i-1]*X[i-1]*X[i-1]
    if(len(X)%2==0):
            for i in range(1,int(len(X)/2+1)):
                if i%2==0:
                    X[i-1],X[len(X)-i+1] = X[len(X)-i+1], X[i-1]
            
            return np.flip(np.r_[B,X],0)
    else:
            return np.flip(np.r_[B, np.flip(X,0)],0)
        
A=np.array([1,2,3,4,5,6,7])
print(transformation(A))
```

```python
def stupid_transformation(X,a=1):                      #Задача №3 решение без NumPy
    B=X.copy()
    
    for i in range(1,len(X)+1):
            if i%2==1:
              X[i-1]=a
            else:
                X[i-1]=X[i-1]*X[i-1]*X[i-1]
    if(len(X)%2==0):
            for i in range(1,int(len(X)/2+1)):
                if i%2==0:
                    X[i-1],X[len(X)-i+1] = X[len(X)-i+1], X[i-1]
            
            return list(reversed(B+X))
    else:
            return list(reversed(B+list(reversed(X))))
        
A=[1,2,3,4,5,6,7]
print(stupid_transformation(A))

```
